```python
# What is the trend in temperature over time?

plt.figure(figsize = (15, 5))
sns.lineplot(x = "Date/Time", y = "Temp_C", data = df)

plt.title("Temperature Over Time")
plt.show()
```

```python
# What is the correlation between temperature and humidity?

plt.figure(figsize = (15, 5))
sns.scatterplot(x = "Rel Hum_%", y = "Temp_C", data = df)
plt.title("Temperature vs Humidity")
plt.show()
```

```python
# What is the distribution of wind speed?

plt.figure(figsize = (15, 5))
sns.histplot(df["Wind Speed_km/h"], bins = 20, kde = True)
plt.title('Wind Speed Distribution')
plt.xlabel('Wind Speed (km/h)')
plt.show()
```

```python
# Is there a relationship between wind speed and visibility?

sns.scatterplot(x = "Visibility_km", y = "Wind Speed_km/h", data = df)
plt.show()
```

```python
# What is the maximum and minimum visibility recorded?

max_visibility = df['Visibility_km'].max()
min_visibility = df['Visibility_km'].min()
print(f'Max Visibility is: {max_visibility} km')
print(f'Min Visibility is: {min_visibility} km')
```

```python
# What is the range of pressure values observed?

plt.figure(figsize=(15,5))
sns.histplot(df['Press_kPa'], bins=20, kde=True)
plt.title('Pressure Distribution')
plt.xlabel('Pressure (kPa)')
plt.show()
```

```python
# Which weather conditions occur most frequently?

plt.figure(figsize=(15,5))
sns.countplot(x = "Weather", data = df, order = df['Weather'].value_counts(ascending = True).index)
plt.title('Most Common Weather Conditions')
plt.xticks(rotation=90)
plt.show()
```
